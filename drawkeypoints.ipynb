```python
import matplotlib.pyplot as plt
import cv2
import os

img = cv2.imread("icones/seta.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.show()
```

```python
detector = cv2.xfeatures2d.SIFT_create()
keypoints, descriptors = detector.detectAndCompute(img, None)
outImage = cv2.drawKeypoints(img, keypoints, None)
plt.imshow(outImage)
plt.show()
```

```python
for file in os.listdir("icones"):
    img = cv2.imread("icones/"+file)
    keypoints, descriptors = detector.detectAndCompute(img, None)
    outImage = cv2.drawKeypoints(img, keypoints, None)
    cv2.imwrite("imgs_with_kps/"+file,outImage)
```
