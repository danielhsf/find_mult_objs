```python
#Haar Cascade
'''
Pros:
    Robusto
    ferramental disponivel pra treinando do arquivo xml disponivel no proprio opencv
Contras:
    Treinando prvio nescessrio para criar um arquivo xml
'''

import cv2
from matplotlib import pyplot as plt

# Inicializa o Classificador Cascade

faceCascade = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")

# Converte o frame para escalas de cinza e descarta os dados de cores

frame = cv2.imread("trump.jpg", 0)
frame_show = frame

# Detecta objetos e retorna o array de objetos

objetos = faceCascade.detectMultiScale( frame, scaleFactor=1.4, minNeighbors=1, minSize=(30, 30), 
                                               flags=cv2.CASCADE_SCALE_IMAGE)

# Desenha um retngulo em torno do objeto

for (x, y, w, h) in objetos:

    cv2.rectangle(frame_show, (x, y), (x+w, y+h), (0, 0, 255), 2)

# Mostra o Frame ao usurio
plt.imshow(frame_show,cmap = 'gray')
plt.show()
```

```python
#Template Matching
#https://docs.opencv.org/2.4/doc/tutorials/imgproc/histograms/template_matching/template_matching.html
'''
Pros:
    implementao fcil
Contras:
    dificuldade em geral o match com diferentes angulos
'''

from __future__ import print_function
import sys
import cv2 as cv
from matplotlib import pyplot as plt

img = cv.imread("/home/daniel/find_mult_objs/Template_Matching_Original_Image.jpg", cv.IMREAD_COLOR)
templ = cv.imread("/home/daniel/find_mult_objs/Template_Matching_Template_Image.jpg", cv.IMREAD_COLOR)

fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.imshow(img,cmap = 'gray')
ax2.imshow(templ,cmap = 'gray')
plt.show()

img_display = img.copy()

result = cv.matchTemplate(img, templ, cv.TM_CCOEFF_NORMED)

_minVal, _maxVal, minLoc, maxLoc = cv.minMaxLoc(result, None)
matchLoc = maxLoc
cv.rectangle(img_display, matchLoc, (matchLoc[0] + templ.shape[0], matchLoc[1] + templ.shape[1]), (0,255,0), 2, 8, 0 )

plt.imshow(img_display,cmap = 'gray')
plt.show()
```
